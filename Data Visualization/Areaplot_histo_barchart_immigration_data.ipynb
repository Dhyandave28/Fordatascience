```python
import numpy as np 
import pandas as pd
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use('ggplot')
```

<h1> Importing data from dataset</h1>

```python
df_can = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv')

print('Data read into a pandas dataframe!')
```

<h1>Displaying first five terms</h1>

```python
df_can.head(5)
```

<h1>Finding dataset entries</h1>

```python
print(df_can.shape)
```

<h1>Setting country name as index</h1>

```python
df_can.set_index('Country', inplace = True)
df_can.head()
```

```python
print('Data dimensions:',df_can.shape)
```

<h1> Creating list of years </h1>

```python
years = list(map(str,range(1980,2014)))
years
```

<h1>Sorting data and Visualizing as areaplot </h1>

```python
df_can.sort_values(['Total'], ascending = False,axis =0,inplace= True)

df_top5 = df_can.head()

df_top5 = df_top5[years].transpose()
df_top5.head()
```

```python
df_top5.index = df_top5.index.map(int)
df_top5.plot(kind='area',stacked = False, figsize=(20,10))

plt.title('Immigration trend of top 5 countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

<h1>To plot unstacked plot (using alpha parameter)</h1>

```python
df_top5.plot(kind ='area',alpha = 0.25,stacked = False , figsize =(20,10))

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

<h1> Plotting using Axes Instance from matplotlib</h1>

```python
ax = df_top5.plot(kind = 'area', alpha = 0.35, figsize =(20,20))

ax.set_title('Immigration trend of top 5 countries')
ax.set_ylabel('Number of immigrations')
ax.set_xlabel('Years')
plt.show()
```

<h1>Using "scripting layer" to creted stacked area plot with "least" immigration</h1>

```python
df_least5 = df_can.tail(5)

df_least5 = df_least5[years].transpose()
df_least5.head()

df_least5.index = df_least5.index.map(int)
df_least5.plot(kind = 'area', alpha = 0.45,figsize =(30,10))
plt.title('Immigration Trend of 5 Countries with Least Contribution to Immigration')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

<h1> Using "Artist layer" to plot unstacked area plot of top 5 countries that contributed "least" to immigration</h1> 

```python
ax = df_least5.plot(kind = 'area', alpha = 0.55,stacked = False, figsize =(20,10))
ax.set_title('Immigration Trend of 5 Countries with Least Contribution to Immigration')
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('Years')
plt.show()
```

<h1>Using numpy to plot histogram</h1>

```python
#To get bin ranges and ferequency
df_can['2013'].head()
```

```python
count, bin_edges = np.histogram(df_can['2013'])
print(count)
print(bin_edges)
```

<h1> Passing xtickes that contain list of bin size</h1>

```python
count,bin_edges = np.histogram(df_can['2013'])

df_can['2013'].plot(kind='hist', figsize=(8, 5), xticks=bin_edges)

plt.title('Histogram of Immigration from 195 countries in 2013') # add a title to the histogram
plt.ylabel('Number of Countries') # add y-label
plt.xlabel('Number of Immigrants') # add x-label

plt.show()
```

<h1>Immigration distribution for denmark, norway and swedenand generating histogram</h1>

```python
df_can.loc[['Denmark', 'Norway', 'Sweden'], years]
```

```python
df_can.loc[['Denmark','Norway','Sweden'],years].plot.hist()
plt.show()
```

<h1> Making the plot look better</h1>

```python
df_t = df_can.loc[['Denmark', 'Norway', 'Sweden'], years].transpose()
df_t.head()
```

```python
df_t.plot(kind='hist', figsize=(10, 6))

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

<h1>Refining the plot</h1>

```python
count, bin_edges = np.histogram(df_t,15)

df_t.plot(kind = 'hist', figsize=(10,6), bins =15, alpha =0.6,xticks = bin_edges,color=['Coral','darkslateblue','mediumseagreen'])

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

<h1>Removing overlap of plot using stacked</h1>

```python
count , bin_edges = np.histogram(df_t,15)
xmin = bin_edges[0] -10
xmax = bin_edges[-1]+10

df_t.plot(kind = 'hist',figsize=(10,6),bins =15,xticks = bin_edges,color=['coral','darkslateblue','mediumseagreen'],stacked = True, xlim = (xmin,xmax))

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants') 

plt.show()
```

<h1>Using scripting layer to display immigration distribution for greece,albania,bulgaria </h1>

```python
df_cof = df_can.loc[['Greece','Albania','Bulgaria'],years]
df_cof = df_cof.transpose()
count , bin_edges = np.histogram(df_cof,15)
df_cof.plot(kind = 'hist', figsize=(10,6),bins = 15,alpha = 0.35,xticks = bin_edges,color =['coral','darkslateblue','mediumseagreen'])

plt.title('Histogram of Immigration from Greece, Albania, and Bulgaria from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

<h1>Bar charts</h1>

<h1>Comparing iceland immigrants </h1>

```python
df_iceland = df_can.loc['Iceland',years]
df_iceland.head()
```

```python
df_iceland.plot(kind = 'bar', figsize=(10,6))

plt.xlabel('Year') 
plt.ylabel('Number of immigrants') 
plt.title('Icelandic immigrants to Canada from 1980 to 2013')

plt.show()
```

<h1>Using the annote method to make plot better </h1>

```python
df_iceland.plot(kind = 'bar', figsize=(10,6),rot =90)

plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.title('Icelandic Immigrants to Canada from 1980 to 2013')

plt.annotate('',
           xy=(32,70),
           xytext =(28,20),
           xycoords='data',
           arrowprops = dict(arrowstyle = '->',connectionstyle ='arc3',color = 'blue',lw=2)
           )

plt.show()
```

```python
df_iceland.plot(kind ='bar', figsize =(10,6), rot = 90)

plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.title('Icelandic Immigrants to Canada from 1980 to 2013')

plt.annotate('',
            xy = (32,70),
            xytext =(30,20),
            xycoords = 'data',
            arrowprops = dict(arrowstyle='->',connectionstyle ='arc3',color ='red',lw=2))
plt.annotate('2008 - 2011 Financial Crisis',  
             xy=(28, 30), 
             rotation=72.5, 
             va='bottom', 
             ha='left', 
             )

plt.show()
```
