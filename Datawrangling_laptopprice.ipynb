```python
import requests

def download(url, filename):
    response = requests.get(url)
    if response.status_code == 200:
        with open(filename, "wb") as f:
            f.write(response.content)
    else:
        print(f"Failed to download: {response.status_code}")
```

```python
avg_weight = df['Weight_kg'].astype('float').mean(axis=0)
df["Weight_kg"].replace(np.nan, avg_weight)
```

```python
common_screen_size = df['Screen_Size_cm'].value_counts().max()
df["Screen_Size_cm"].replace(np.nan,common_screen_size)
```

```python
df["Weight_kg"]= df["Weight_kg"].astype('float')
df["Screen_Size_cm"]= df["Screen_Size_cm"].astype('float')
```

```python
df["Weight_kg"]= df["Weight_kg"]/2.205
df.rename({'Weight_kg': 'Weight_pounds'},inplace = True)

df["Screen_Size_cm"] = df["Screen_Size_cm"]/2.54
df.rename({'Screen_Size_cm': 'Screen_Size_Inches'},inplace = True)
```

```python
df['CPU_frequency'] = df['CPU_frequency']/df['CPU_frequency'].max()
```

```python
bins = np.linspace(min(df["Price"]), max(df["Price"]),4)
group_names = ['low', 'medium','high']
df["Price-binned"] = pd.cut(df['Price'], bins, labels=group_names, include_lowest = True)
```

```python
plt.bar(group_names, df["Price-binned"].value_counts())
plt.xlabel("Price")
plt.ylabel("count")
plt.title("Price bins")
```

```python
dummy_variable1 = pd.get_dummies(df['Screen'])
dummy_variable1.rename({'IPS panel': 'Screen-IPS_panel','Full HD': 'Screen-Full_HD'}, inplace = True)
df = pd.concat([df, dummy_variable1], axis =1)

df.drop("Screen", axis = 1, inplace = True)
```
